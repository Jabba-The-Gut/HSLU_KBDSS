### Aufgabe 1 N-Grams

```python
#Testat Aufgabe 1 - N-Grams
import nltk, re, string
from nltk.util import ngrams
from nltk.corpus import stopwords, gutenberg
import itertools

# download gutenberg
nltk.download('gutenberg')

# dowload stopwords
nltk.download('stopwords')

# load gutenberg corpus
text = gutenberg.raw('shakespeare-macbeth.txt')
```

```python
# Text cleanup
text = re.sub('<.*>', '', text) # remove xml markup
text = re.sub('ENDOFARTICLE.', '', text) # ENDOFARTICLE entfernen
text = re.sub("[" + re.sub("\.","",string.punctuation) + "]", "", text) # Satzzeichen entfernen
```

```python
# Tokenize text (split in words)
tokenized = text.split()
```

```python
# Remove stopwords (words with no meaning like I or my)
stop_words = set(stopwords.words('english'))
tokenized = [w for w in tokenized if not w in stop_words]
```

```python
# Count bigrams
bigrams = nltk.bigrams(tokenized)
freq_bigrams = nltk.FreqDist(bigrams)
freq_dist = nltk.FreqDist(freq_bigrams)
print(f'Found {len(freq_dist)} bigrams')
```

```python
# print bigrams
bigrams = dict(itertools.islice(freq_dist.items(), 10))

for bigram in bigrams:
    print(bigram)
```

```python
# Count trigrams
trigrams = nltk.trigrams(tokenized)
freq_trigrams = nltk.FreqDist(trigrams)
freq_dist = nltk.FreqDist(freq_trigrams)
print(f'Found {len(freq_dist)} trigrams')
```

```python
# print trigrams
trigrams = dict(itertools.islice(freq_dist.items(), 10))  

for trigram in trigrams:
    print(trigram)
```

### Aufgabe 2 TF-IDF

```python
#Testat Aufgabe 2 - TF-IDF
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer

with open('IMDB_1.txt', 'r') as myfile:
    imdb_1=myfile.read().replace('\n', '')

with open('IMDB_2.txt', 'r') as myfile:
    imdb_2=myfile.read().replace('\n', '')

with open('IMDB_3.txt', 'r') as myfile:
    imdb_3=myfile.read().replace('\n', '')



imdb_text = [
    imdb_1,
    imdb_2,
    imdb_3
]

documents_names = ["IMDB_1", "IMDB_2", "IMDB_3"]

vect = TfidfVectorizer(stop_words='english', ngram_range=(2,2))
tfidf = vect.fit_transform(imdb_text).todense()
pd.DataFrame(tfidf, columns=vect.get_feature_names(), index=documents_names).T
```
